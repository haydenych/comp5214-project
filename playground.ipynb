```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
im1 = Image.open("./data/vimeo_triplet/sequences/00001/0001/im1.png")
im2 = Image.open("./data/vimeo_triplet/sequences/00001/0001/im2.png")
im3 = Image.open("./data/vimeo_triplet/sequences/00001/0001/im3.png")
```

```python
f, axarr = plt.subplots(3,1, figsize=(5,12))
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.1)

axarr[0].imshow(im1)
axarr[1].imshow(im2)
axarr[2].imshow(im3)

axarr[0].set_title("Frame 0")
axarr[1].set_title("Frame 1")
axarr[2].set_title("Frame 2")
```
